User Based Filtering

```python
user_m = pd.pivot_table(df,index="user_id",columns="book_title",values="book_rating")
```

```python
user_m
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(user_m.fillna(0))
# replace the Nun value = 0
```

```python
similarity
```

```python
us = pd.DataFrame(similarity,index=user_m.index,columns=user_m.index)
```

```python
us.head()
```

```python
np.fill_diagonal(us.values,0)
# This ensures that when finding the most similar user, a user will not be considered as their own most similar user
#so the similarity of user with their own is set 0
```

```python
us.head()
```

```python
pair = us.idxmax()
#pair1 = us.idxmax(axis=1)[:10] this is used for the identify columns
```

```python
pair.head(10)
# the first 10 maximum value occurs along each columns
```

```python
sim_user_df=pd.DataFrame(pair,columns=['similar_user'])
sim_user_df.iloc[0:10]
# most highest similiraty between user to user
```

Item Based Filter

```python
item = user_m.T
```

```python
item.head()
```

```python
item.fillna(0,inplace=True)
```

```python
similarity_score=cosine_similarity(item)
cosine_score_df=pd.DataFrame(similarity_score,index=item.index,columns=item.index)
cosine_score_df
# find the cossine similarity of item vs item
```

```python

np.fill_diagonal(cosine_score_df.values, 0)
```

```python
sim_books=cosine_score_df.idxmax(axis=1)
# columns label contain the maximum value along each row idxmax maxium index of data frame
```

```python
sim_books.head(10)
#this is most similar book
```

```python
similar_book = pd.DataFrame(sim_books)
```
